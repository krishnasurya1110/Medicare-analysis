```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("drugs_data.csv")
```

```python
data.head()
```

```python
data = data[data['Mftr_Name'] != 'Overall']
data.head()
```

```python
data.drop(columns=['Tot_Mftr'], inplace=True)
```

```python
data.columns
```

```python


# List of years to process
years = range(2017, 2022)
new_data_list = []

# Loop through each year to create a dataframe with uniform column names
for year in years:
    # Columns specific to each year
    year_columns = [
        f'Tot_Spndng_{year}', f'Tot_Dsg_Unts_{year}', f'Tot_Clms_{year}', f'Tot_Benes_{year}',
        f'Avg_Spnd_Per_Dsg_Unt_Wghtd_{year}', f'Avg_Spnd_Per_Clm_{year}', f'Avg_Spnd_Per_Bene_{year}',
        f'Outlier_Flag_{year}'
    ]
    
    # Check which of these columns actually exist in the data
    existing_columns = [col for col in year_columns if col in data.columns]

    # Create a subset dataframe for the year
    df_year = data[['Brnd_Name', 'Gnrc_Name', 'Mftr_Name'] + existing_columns].copy()
    
    # Standardize column names by removing the year suffix
    rename_dict = {col: col.split('_' + str(year))[0] for col in existing_columns}
    df_year.rename(columns=rename_dict, inplace=True)
    
    # Add the year column
    df_year['Year'] = year
    
    # Append to list
    new_data_list.append(df_year)

# Concatenate all the dataframes into a single one
consolidated_data = pd.concat(new_data_list, ignore_index=True)

# Fill 0 in place of NaN for numerical columns
numerical_cols = consolidated_data.select_dtypes(include='number').columns
consolidated_data[numerical_cols] = consolidated_data[numerical_cols].fillna(0)
consolidated_data.head()

# Rearrange columns
consolidated_data = consolidated_data[['Brnd_Name', 'Gnrc_Name', 'Mftr_Name', 'Year', 'Tot_Spndng', 'Tot_Dsg_Unts', 'Tot_Clms', 'Tot_Benes', 'Avg_Spnd_Per_Dsg_Unt_Wghtd', 'Avg_Spnd_Per_Clm', 'Avg_Spnd_Per_Bene', 'Outlier_Flag']]

# Save the transformed data to a new CSV file
consolidated_data.to_csv('drugdata_transformed.csv', index=False)
```
